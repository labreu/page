```python
import pandas as pd
import io
import urllib, base64

%pylab inline

def plot_to_html(f, filename='plt.html', save=False):
    buf = io.BytesIO()
    f.savefig(buf, format='png')
    buf.seek(0)
    string = base64.b64encode(buf.read())

    uri = 'data:image/png;base64,' + urllib.parse.quote(string)  # urllib.quote for python2
    html = '<img src = "%s"/>' % uri
    if save:
        with open(filename, 'w+') as f:
            f.write(html)
        return
    else:
        return html
```
